# Import Libraries

```python
#Working With DataFrame
import pandas as pd

# Python Structs Type For Wide Variety Mathematical Operations Like Arrays
import numpy as np

# Libraries For Createing Visualizations
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Load The CSV To DataFrame
dataset = pd.read_csv("BitData_Final.csv")
```

# Google's Trends 

```python
# Use Bitcoin Low VS Hig Values Plots
plt.figure(figsize = (12, 7))
dataset[0:365].plot(x='DATE', y=['TRENDS MEAN'], figsize=(10,5), grid=True)
plt.title("GOOGLE Trends This Year", size=25)
```

# Low_High Visualizations 

```python
# Use Bitcoin Low VS Hig Values Plots
plt.figure(figsize = (12, 7))
dataset[0:365].plot(x='DATE', y=['LOW', 'HIGH'], figsize=(10,5), grid=True)
plt.title("Low VS High Year", size=25)
```

# Twitter's Visualizations   

```python
# Caculating For The Tweets PIE

sum_p = sum_n=0
for i in dataset["POSITVE ANALZE TWEETS"]:
    sum_p = sum_p + i
for i in dataset["NEGATIVE ANALZE TWEETS"]:
    sum_n = sum_n + i

y = np.array([sum_p, sum_n])
mylabels = ["Postive", "Negative"]

# Making Tweets Plot
plt.title("Tweets",size=25)
plt.pie(y, labels = mylabels, startangle = 90)
plt.show() 
```

```python
# Making Tweets Plot
list1 = []
for i in range(0,50):
    list1.append(i)

namep = dataset["POSITVE ANALZE TWEETS"].head(50)
namen = dataset["NEGATIVE ANALZE TWEETS"].head(50)
barWidth = 0.20
br1 = np.arange(len(namen))
br2 = [x + barWidth for x in br1]

# Figure Size
fig = plt.subplots(figsize =(20, 5))
 
# Horizontal Bar Plot
plt
plt.bar(br1, namep, width = barWidth, color = "blue")
plt.bar(br2, namen, width = barWidth, color = "yellow")
plt.xticks([r + barWidth for r in range(len(namen))],list1)
 
# Adding Xticks
plt.title("Tweets Anaylsis", fontweight ='bold', fontsize = 25)
plt.xlabel('Days Count', fontweight ='bold', fontsize = 15)
plt.ylabel('Number Of Tweets', fontweight ='bold', fontsize = 15)

plt.legend(mylabels)
 
# Show Plot
plt.show()
```
